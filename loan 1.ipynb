## 9.      Write function for find ‘funded_amt’ level . if ‘funded_amt’ <=5000  consider as -1 if ‘funded_amt’ >5000 and <= 15000  consider as 0 if ‘funded_amt’ >15000  consider
as 1

```python
def fun_amt(data):
    data.loc[data['funded_amnt'] <=5000, 'funded_amnt'] = -1
    data.loc[(data['funded_amnt']>5000) & (data['funded_amnt']<=15000), 'funded_amnt']=0
    data.loc[data['funded_amnt'] >15000, 'funded_amnt'] = 1
    
    return data

fun_amt(data)
```
